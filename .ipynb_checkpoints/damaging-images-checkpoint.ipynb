# Damaging Images

This notebook develops our methods for damaging images, with a number of different methods simulating different forms of damage that may be encountered in real applications (particularly in medical imaging).

These methods are as follows:
- **Corruption (random)**: Randomly selecting pixels to be zeroed, which simulates the corruption of individuals pixels over a random distribution.
- **Corruption (patchy)**: Randomly selecting regions ("patches") to be zeroed, which simulates a more structured corruption of data, though still over a random distribution.
- **Noise (random)**: Similar to corruption (random), but introducing random noise rather than zeroing.
- **Noise (aggregated)**: Likewise, but a pixel's noise is sampled from its immediate neighbourhood (e.g. aggregate pixel value in the local region) to simulate confounding noise.
- **Gaussian blur**: Applying a Gaussian blur across the image to simulate a wide-spread muting of information (i.e. not a complete loss, but a loss of resolution at least).
- ...

Each method is to be parameterised by the `damage_degree`, which exists as a value in $(0,1)$ describing how much damage the image has been exposed to. It will have different interpretations under different methods (e.g. in random corruption it describes the proportion of pixels exposed to corruption, in Gaussian blur it controls the spread of information over the image, etc.).

## Corruption (Random)

This method involves randomly sampling pixels in the image to be 'corrupted'. Corruption amounts to zeroing the value -- the data is entirely lost.

We take no mitigating steps here, since this is something we expect the model to learn automatically or otherwise. So, while it may be advisable to fill corrupted pixels with some aggregation of their neighbour's values, we will not take such an approach (though, exactly this is considered as "aggregated noise" damage later).

Here, `damage_degree` translates to the percentage of pixels in the image that are to be corrupted. This is deterministic -- once a pixel is selected for corruption, it *is* corrupted. We may consider a stochastic corruption method in the future to reduce predictability (in that the model is theoretically able to anticipate how much damage any given image will have), but we will not in this method.

```python
def apply_random_corruption(img, damage_degree):
    """
    Applies random pixel corruption to an image.

    Parameters:
        img (PIL.Image): The image to corrupt.
        damage_degree (float): The percentage of pixels to corrupt (0 to 100).
    
    Returns:
        PIL.Image: The corrupted image.
    """
    # Convert the image to a PyTorch tensor
    transform = transforms.ToTensor()
    img_tensor = transform(img)

    # Calculate the total number of pixels to corrupt based on the damage_degree
    total_pixels = img_tensor.numel()
    pixels_to_corrupt = int(total_pixels * damage_degree / 100)

    # Generate random indices for selecting pixels to corrupt
    indices = torch.randperm(total_pixels)[:pixels_to_corrupt]

    # Flatten the image tensor to easily use the linear indices
    flat_img_tensor = img_tensor.view(-1)

    # Corrupt the selected pixels by setting their values to zero
    flat_img_tensor[indices] = 0

    # Reshape the tensor back to its original shape
    corrupted_img_tensor = flat_img_tensor.view_as(img_tensor)

    # Convert the tensor back to a PIL Image for display or saving
    to_pil_image = transforms.ToPILImage()
    corrupted_img = to_pil_image(corrupted_img_tensor)

    return corrupted_img
```

```python
# Apply damage effect
damaged_img = apply_random_corruption(example_image, damage_degree=50)
damaged_img_np = np.array(damaged_img)

plt.imshow(damaged_img_np)
plt.title('Damaged Image')
plt.axis('off')  
plt.show()
```

## Corruption (patchy)

```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import numpy as np

def apply_patchy_corruption(img, damage_degree, patch_size=20):
    """
    Applies patchy corruption to an image based on a damage degree.

    Parameters:
        img (PIL.Image): The image to corrupt.
        damage_degree (float): The percentage of the image area to be covered by patches.
        patch_size (int, optional): The size of each square patch (in pixels). Default is 20.
    
    Returns:
        PIL.Image: The corrupted image.
    """
    # Convert the image to a PyTorch tensor
    transform = transforms.ToTensor()
    img_tensor = transform(img)

    # Image dimensions
    _, H, W = img_tensor.shape

    # Total image area and area to be corrupted
    total_area = H * W
    corrupt_area = total_area * damage_degree / 100

    # Calculate the number of patches needed to cover the desired area
    patch_area = patch_size ** 2
    num_patches = int(np.ceil(corrupt_area / patch_area))

    for _ in range(num_patches):
        # Randomly select the top-left corner of the patch
        x = np.random.randint(0, W - patch_size)
        y = np.random.randint(0, H - patch_size)

        # Zero out the pixels in the selected patch
        img_tensor[:, y:y+patch_size, x:x+patch_size] = 0

    # Convert the tensor back to a PIL Image
    to_pil_image = transforms.ToPILImage()
    corrupted_img = to_pil_image(img_tensor)

    return corrupted_img
```

```python
# Apply damage effect
damaged_img = apply_patchy_corruption(example_image, damage_degree=30)
damaged_img_np = np.array(damaged_img)

plt.imshow(damaged_img_np)
plt.title('Damaged Image')
plt.axis('off')  
plt.show()
```

## Adding Blur


```python
def apply_damage_effect(img, damage_degree, filter_type='circle'):
    transform = transforms.ToTensor()
    
    img_tensor = transform(img).unsqueeze(0)  # Add batch dimension

    ksize = max(3, min(21, 3 + damage_degree * 2))  # Example way to use damage_degree
    kernel = torch.zeros((ksize, ksize))
    center = ksize // 2

    if filter_type == 'circle':
        for i in range(ksize):
            for j in range(ksize):
                if (i - center) ** 2 + (j - center) ** 2 <= (ksize // 3) ** 2:
                    kernel[i, j] = 1
    elif filter_type == 'horizontal':
        kernel[center, :] = 1
    elif filter_type == 'vertical':
        kernel[:, center] = 1
    elif filter_type == 'diagonal':
        torch.fill_diagonal_(kernel, 1)
    else:
        raise ValueError("Invalid filter type. Choose 'circle', 'horizontal', 'vertical', or 'diagonal'.")

    # Normalize the kernel
    kernel /= kernel.sum()

    # Apply the kernel to the image
    kernel = kernel.expand(img_tensor.size(1), 1, ksize, ksize)  # Adjust kernel shape for convolution
    padding = ksize // 2
    convolved = torch.nn.functional.conv2d(img_tensor, kernel, padding=padding, groups=img_tensor.size(1))

    # Convert tensor back to image
    to_pil_image = transforms.ToPILImage()
    damaged_img = to_pil_image(convolved.squeeze(0))

    return damaged_img
```

```python
# Apply damage effect
damaged_img = apply_damage_effect(example_image, damage_degree=50, filter_type='vertical')
damaged_img_np = np.array(damaged_img)

plt.imshow(damaged_img_np)
plt.title('Damaged Image')
plt.axis('off')  
plt.show()
```
